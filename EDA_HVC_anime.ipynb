```python
#Importo as librerías
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Abro o ficheiro obtido de https://www.kaggle.com/vishalmane10/anime-dataset-2022?select=Anime.csv
anime = pd.read_csv('Anime.csv', index_col = 0);
```

```python
anime.info()
```

```python
#Elimino as columnas con un porcentaxe superior ao 50 % de nulos ademais da columna Related_anime que non ten demasiada importancia, ao igual que a columna Voice_actors
anime = anime.dropna(axis=1,thresh=(len(anime)/2))
anime = anime.drop('Related_anime', axis=1)
anime = anime.drop('Voice_actors', axis=1)
anime.shape
```

```python
#Trato individualmente las otras columnas
#Elimino algún espazos que hai no tipo
anime.Type = anime.Type.apply(lambda x: x.strip())
anime['Description'] = anime[['Description']].apply(lambda x: x.fillna('ningunha'),axis=0)
anime['Release_year'] = anime[['Release_year']].apply(lambda x: x.fillna(x.mean()),axis=0)
anime['Tags'] = anime[['Tags']].apply(lambda x: x.fillna('descoñecido'),axis=0)

anime.isna().sum()
```

```python
# Buscar filas duplicadas
print(f'Total de filas duplicadas: {anime.duplicated().sum()}')
```

```python
#Borramos as filas duplicadas
anime.drop_duplicates(inplace=True)
anime.shape
```

```python
#O resto de nulos deixoos e gardoo unha copia
anime.to_csv('anime_limpo.csv',index=False)
```

```python
#VISUALIZACIÓN
sns.displot(anime.Rating, height=7,kde=True)
plt.title("Distribución de la valoración", size=18);
#A maioría dos animes teñen unha valoración comprendida entre 3 e 4 puntos, moi poucos consiguen unha puntuación maior e al menos
# neste dataset ningún anime recibe a valoración más alta(5), tamén apréciase que moi poucos animes reciben unha valoración excesivamente baixa
```

```python
sns.histplot(anime.Release_year, color="red", kde=True)
plt.title('Distribución de su año de salida', size=18);
#Neste gráfico e tendo en conta que apliquei aa media para sustituir os valores nulos deste campo
#apréciase que a produción dos anime tivo un gran auxe a partir do 1980, alcanzando un gran pico entorno ao 2012 e que a día de hoxe segue
#mantendo un nivel alto, ademáis vense que os primeros animes surxiron a principios do 1900, hace 100 anos
```

```python
sns.displot(anime.Episodes, height=10, color="black")
plt.title('Distribución del número de episodios', size=18);
#Pódese apreciar que os animes que máis destacan son os cortos, cunha cantidade de episodios comprendida principalmente entre os 12 e 24
#Ademáis apréciase que os animes con máis de 100 episodios vólvense moi escasos, e vanse facendo máis escasos conforme acadan un número de episodios elevados
```

```python
#Esta é a forma das distribucións do tipo de anime
plt.figure(figsize=(15,15))
sns.barplot(y=anime['Type'], x=anime['Type'].index, ci=None)
plt.title('Distribution of {}'.format('Type'), size=20)
plt.show()
#Apréciase que maioritariamente os animes ocupan diversos formatos, poden ser sacados en internet ao mesmo tempo
#que se emiten por televisión ou se distribúen mediante dvds, aínda que non comprendo o porque de na forma de distribución se inclúe o 'tipo de contenido' (movie, music, OVA)
```

```python
plt.title('Influencia do tipo de anime e a valoración', size=20)
sns.boxplot(x=anime.Type, y=anime.Rating, palette='rainbow');
```

```python
top_studios = anime.Studio.value_counts()[:15]
sns.barplot(y=top_studios.index, x=top_studios, ci=False)
plt.title('Estudios de anime con su respectivas cantidades', size=20);
#O estudo que máis animes produciu foi TOEI Animation por unha gran diferenza
```

```python
studios_more_20 = list(anime.Studio.value_counts().index[:111]) #Os 111 primeiros estudos animaron mínimo 20 animes
top_studios_rtg = anime[anime.Studio.isin(studios_more_20)].groupby('Studio').Rating.mean().sort_values(ascending=False)[:15]
plt.figure(figsize=(15,15))
sns.barplot(y=top_studios_rtg.index, x=top_studios_rtg.values).set_title('Valoración dos mellores estudios que posúen máis de 20 animes', size=20);
```
